# Series02-Python : Sinusoidal Waves

the equation of a sinusoidal wave is given by : 

Y(t) = A*sin(wt + phi)
where : 
y(t) : is the displacement at time
A : the amplitude
w : is the angular frequency
phi : the phase shift 
t : is time


```python
import numpy as np 
import matplotlib.pyplot as plt 

# parameters 
A = 1
T = 1
w = 2 * np.pi/T
phi = 0

# Time array 
n_periods = 1 # Number of periods to display
duration = n_periods*T # duration in second
t = np.linspace(0,duration,1000) 

y = A * np.sin(w*t+phi)
z = A * np.cos(w*t+phi)
# add a second axis y for the cos
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(t,y, 'b-', label ='sine wave')
ax2.plot(t,z,'r-', label = 'Cosine waves')
# add a second label to the y axis
ax2.set_ylabel('cosine wave Amplitude', color = 'red')
ax1.set_ylabel('Sine wave Amplitude', color = 'blue')
ax1.legend()

plt.plot(t,z, color= ('red'))
plt.plot(t,y, color=('blue'))
plt.title('Sinusoidal wave')
plt.xlabel('time (s)')
plt.legend()
plt.grid()
plt.show()
```

# create a function that generate a sinusoidal wave 

```python
# function

def f_sin(Amplitude,time,pulsation,phase):
    y = Amplitude * np.sin(pulsation*time+phase)
    return y 
# parameters 
a = 10
frequency = 1
start_time = 0
stop_time = 5
n_samples = 1000
t = np.linspace(start_time, stop_time, n_samples)
w = 2 * np.pi * frequency 
phi = 0
sinus_signal = f_sin(a,t,w,phi)



plt.plot(t,sinus_signal)
plt.title('Sinusoidal Wave')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (y)')
plt.grid()
plt.show()
```
